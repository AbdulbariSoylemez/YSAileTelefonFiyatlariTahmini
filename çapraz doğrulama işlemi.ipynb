```python
import numpy as np 
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import StratifiedKFold
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
```

```python
### kodu inceleyerek çapraz doğrulama işlemi yapalım ve buna göre başarımı değerlendiriniz
```

```python

veri = pd.read_csv("/Users/abdulbarisoylemez/Documents/Visual Code/YapayZekaDersi/3.hfta/archive/train.csv")

# Etiket kodlaması yapalım
label_encoder = LabelEncoder()
labels = label_encoder.fit_transform(veri["price_range"])
classes = label_encoder.classes_
```

```python

X = veri.drop(["price_range"], axis=1)
Y = labels

sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python

average_train_loss = []
average_train_accuracy = []
average_val_loss = []
average_val_accuracy = []
```

```python
for train_index, test_index in kfold.split(X, Y):
    x_train, x_test = X[train_index], X[test_index]
    y_train, y_test = Y[train_index], Y[test_index]

    # Etiketleri One-Hot Encoding yapalım
    y_train = to_categorical(y_train)
    y_test = to_categorical(y_test)

    model = Sequential()
    model.add(Dense(16, input_dim=20, activation="relu"))
    model.add(Dense(12, activation="relu"))
    model.add(Dense(12, activation="relu"))
    model.add(Dense(4, activation="softmax"))

    model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
    model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=50)

    average_train_loss.append(np.mean(model.history.history["loss"]))
    average_train_accuracy.append(np.mean(model.history.history["accuracy"]))
    average_val_loss.append(np.mean(model.history.history["val_loss"]))
    average_val_accuracy.append(np.mean(model.history.history["val_accuracy"]))
```

```python
print("Ortalama Eğitim Kaybı:", np.mean(average_train_loss))
print("Ortalama Eğitim Başarımı:", np.mean(average_train_accuracy))
print("Ortalama Doğrulama Kaybı:", np.mean(average_val_loss))
print("Ortalama Doğrulama Başarımı:", np.mean(average_val_accuracy))
```
